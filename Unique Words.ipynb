```python
from collections import Counter
import re

def extract_unique_words(text):
    # Zerlege den Text in Wörter unter Berücksichtigung von Groß-/Kleinschreibung
    words = re.findall(r'\b\w+\b', text)

    # Zähle die Häufigkeit jedes Wortes
    word_counts = Counter(words)

    # Filtere nur Wörter, die genau einmal vorkommen
    unique_words = [word for word, count in word_counts.items() if count == 1]

    # Sortiere die Liste alphabetisch
    unique_words.sort()

    return unique_words

# Beispieltext
text = "ABRATEN von der Mission ABSURDITÄT glaubhaft zu halten VORTÄUSCHEN ABSURDITÄT nachgehen Abtretung der Schönen stehet bevor Adoption AMUSEMENT An einen AUSWEGLOSEN Ort gelangen An Königswahl teilnehmen Änderung der EXISTENZFORM beheben ANEIGNEN beheben Angebliche Testnotwendigkeit VORTÄUSCHEN Angebot GEOPFERT ZU WERDEN Angebot GEOPFERT ZU WERDEN nachgehen Angebot KÖNIG ZU WERDEN Angst mit Angst vergelten Auf dem Heimweg Auf dem VERTRAG bestehen Auf Hilfsquelle zurückgreifen Auferstehung VORTÄUSCHEN Aufklären der ABSURDITÄT Auftrag als ABSURDITÄT Auftrag zum Kochen Auftrag zum Unterhalten Auftrag zum VOTÄUSCHEN Auslösen von FERNLIEBE Ausplaudern AUSRÜSTUNG als ZAUBERMITTEL erhalten Aussetzung Baumersteigen Bauwerk errichten Bedingungslose Loyalität VORTÄUSCHEN Begegnung Begehren einer Person oder eines GEGENSTANDES Beglaubigen der Behauptung der ABSURDITÄT Behauptung der ABSURDITÄT BEHINDERUNG durch einen Fluch BEHINDERUNG durch Verstümmelung Beiwohnung mit der schlafenden Schönen BESITZER DES OBJEKTES erfassen BESITZER DES OBJEKTES herbeiholen Bestrafung Beuteteilen Bis auf Auferstehung ruhen lassen BOTSCHAFTEN verfälschen BOTSCHAFTEN versenden Das Geheimnis des Erwerbs der EXTERNEN SEELE ausspionieren Das Geheimnis des Erwerbs des besonderen GEGENSTANDES ausspionieren Das Geheimnis des Erwerbs des REITTIERs ausspionieren Das VORTÄUSCHEN eingestehen Dem ABRATEN von der Mission nachgehen Dem AMUSEMENT nachgehen Dem Angebot KÖNIG ZU WERDEN nachgehen Dem GEHEIMZEICHEN nachgehen Dem STAUNEN nachgehen Dem VORTÄUSCHEN nachgehen DEMÜTIGUNG der verleumdeten Person Den ALTERNATIVEN nachgehen Den Einbrecher begnadigen Den GIER und NEID nachgehen Den prächtigen Palast erhalten Den STALL ausmisten Den STALL ausmisten in Angriff nehmen Den Vorgang deuten Der ABSURDITÄT entgegentreten Der MISSETATwarnung nachgehen Der NEUGIER nachgehen Der PROVOKATION nachgehen Der RESPEKTLOSIGKEIT entgegentreten Der VERLOCKUNG nachgehen Der VORHERSAGE nachgehen Der Wechsel der IDENTITÄT eingestehen Die Änderung der EXISTENZFORM Die Änderung der EXISTENZFORM stören Die Beschaffung der AUSRÜSTUNG Die ERNTE verarbeiten Die ERNTE verarbeiten in Angriff nehmen Die Hilfsquelle greift zurück Die potenzielle Beute begnadigen Die SCHÖNE erfassen Die SCHÖNE herbeiholen Die VORHERSAGE tritt ein Eigene Misshandlung nachforschen Ein VERSPRECHEN halten Eine Bedrohung VORTÄUSCHEN Eine BEHINDERUNG beheben Eine BLAMAGE bereiten Eine BLAMAGE gelingt Eine DEMÜTIGUNG beheben Eine ENTFÜHRUNG Eine ENTFÜHRUNG beheben Eine Enttäuschung ausgelöst Eine Gefahr VORTÄUSCHEN Eine Heiratsabsicht kundtun Eine löst andere ABSURDITÄT aus Eine menschliche OPFERGABE verlangen Eine numinose Einmischung VORTÄUSCHEN Einem Gemenge entkommen Einem TODESAUFTRAG entkommen Einen Antagonist einstellen Einen Begleiter gewinnen Einen gold oder GELDPRODUZIERENDEN ZAUBERMITTEL erhalten Einen magischen EINGRIFF vorbereiten Einer EINLADUNG nachgehen Eines auffälligen GEGENSTANDES habhaft werden EINLADUNG Ende Enthüllung der Wahrheit durch das subtile Mittel Entsorgen Erlösung Erträumen ERZÄHLEN beenden ERZÄHLEN beginnen EXTERNEN SEELE erfassen EXTERNEN SEELE vernichten Familienzusammenführung Fehlende männliche Nachkommen FEHLENDEN Gegenstand besorgen FEHLENDEN Gegenstand erfassen FELDER BESTELLEN FELDER BESTELLEN in Angriff nehmen FERNLIEBE beheben FLUCH FLUCH abwenden FLUCHT FLUCHT vorbereiten Folge von Arbeitsscheu deuten Folge von Betrügerei deuten Folge von Bigotterie deuten Folge von Geiz deuten Folge von Grobheit deuten Folge von Klatsch deuten Folge von Kriminalität deuten Folge von Vernachlässigung der Eltern deuten Folge von verweigerter Gastfreundschaft deuten Folge von vorehelichem Geschlechtsverkehr deuten Folgen eines TABUbruches rückgängig machen Forderung oder Bitte als ABSURDITÄT FREIHEITSBERAUBUNG FREIHEITSBERAUBUNG beheben Fremdes Gut einziehen Freundschaft schließen FRUCHT besorgen FRUCHT erfassen Gastfreundlichkeit VORTÄUSCHEN Geänderte EXISTENZFORM VORTÄUSCHEN Geburt der Angehörige Gegen ein Tier kämpfen GEGENLEISTUNG anbieten GEGENLEISTUNG FÜR GESCHENK oder Aufmerksamkeit erhalten GEGENLEISTUNG FÜR GESCHENK oder Aufmerksamkeit gewähren GEGENLEISTUNG FÜR GESCHENK oder Aufmerksamkeit nachgehen GEGENLEISTUNG FÜR HILFE gewähren GEHEIME LIEBESBEZIEHUNG GEHEIME LIEBESBEZIEHUNG beheben GEHEIMNIS AUFKLÄREN GEHEIMNIS AUFKLÄREN in Angriff nehmen GEHEIMZEICHEN vermitteln Gerechte Rückgabe fordern Geschick VORTÄUSCHEN Geschlechtsumwandlung erzielen Gespräche belauschen GIER und NEID ausgelöst Goldhaariges Kind zur Welt bringen Große Mengen ESSEN Gutes Benehmen Handeln als ABSURDITÄT Hartes Großes Mächtiges mit Weichem Kleinem Schwachem bewältigen Heilkunst VORTÄUSCHEN HEILMITTEL besorgen HEILMITTEL erfassen Heimlich beobachten Heiraten Heiratszusage erpressen Held erkennen Heldentat VORTÄUSCHEN HERBEIRUFENDES ZAUBERMITTEL erhalten HERDE besorgen HERDE erfassen Hilfe erpressen Hilfeversprechen erhalten Hitze entkommen Hochzeit verschieben Hohlen Gegenstand zum Verstecken erhalten HOLZ besorgen HOLZ Sammeln in Angriff nehmen HUNGERSNOT HUNGERSNOT beheben IDENTITÄT ändern IDENTITÄT beweisen Im Verborgenen lauern Information aus zweiter Hand Intensive ERKUNDUNGSAKTION Intensive ERKUNDUNGSAKTION nachgehen Interesse VORTÄUSCHEN Jemanden in Kannibalismus verwickeln Jenseitswanderung unternehmen JUWELEN besorgen JUWELEN erfassen Kaiserschnitt Kälte entkommen Kinderlosigkeit KINDERMORD erforderlich KINDERMORD nachgehen Knochen begraben Kollaborationsbereitschaft VORTÄUSCHEN Komplikationsfreier Flug KONTAKT abbrechen KONTAKT wiederherstellen Kraft VORTÄUSCHEN KRANKHEIT KRANKHEIT beheben Krankheit VORTÄUSCHEN Krieg führen Krisenursachen ergründen Letzte Gelegenheit nützen Liebe wird ausgelöst Lösung für das LEIDENDE TIER deuten Lösung für das LEIDENDE TIER nachgehen Lösung für den FÄHRMANN deuten Lösung für den FÄHRMANN nachgehen Lösung für HEIRATSWILLIGE deuten Lösung für HEIRATSWILLIGE nachgehen Lösung für SCHATZHÜTER deuten Lösung für SCHATZHÜTERN nachgehen Magische EMPFÄNGNISmittel erfassen Magische EMPFÄNGNISmittel erhalten Magische EMPFÄNGNISmittel suchen MARKIERUNGEN anbringen MARKIERUNGEN erkennen MARKIERUNGEN vorlegen MASSENBEDROHUNG MASSENBEDROHUNG beheben Mehrere Aufgaben in Folge erledigen Mehrere Gegner nacheinander besiegen Mehrlingsgeburt MEINEID erpressen MEINEID erpressen beheben Menschliche Körperteile in Edelmetall verwandeln Menschliche OPFERGABE beheben Menschliches Blut als HEILMITTEL erforderlich MILCH besorgen MILCH erfassen Missachtung der LIEBE Missachtung der LIEBE beheben MISSETAT beheben MISSETAT planen MISSETAT wie Aneignung MISSETAT wie Behinderung MISSETAT wie Betäubung MISSETAT wie ein Surrogat unterschieben MISSETAT wie Entmachtung MISSETAT wie Falle stellen MISSETAT wie Hinterhalt MISSETAT wie Jemanden im Stich lassen MISSETAT wie Kannibalismus MISSETAT wie Tötung MISSETAT wie Vergiftung Missgeburt Mit einer anderen behinderten Person einen Bund schließen MUSIKINSTRUMENT besorgen MUSIKINSTRUMENT erfassen Mut zeigen N Nach dem eigenen VORTÄUSCHEN handeln Nahrungsvorrat aufgebraucht NEUGIER ausgelöst Notwendiges auf wunderbare Weise besorgen NUTZTIER besorgen NUTZTIER erfassen Opfer des eigenen VORTÄUSCHENs werden PARALYSIERUNG PARALYSIERUNG beheben Personenschaden anrichten PFLANZE besorgen PFLANZE erfassen PFÖRTNER besänftigen PFÖRTNER entkommen PFÖRTNER täuschen Potenziellen Feind zum Gastgeber gewinnen Prächtige Kleidung und Schmuck erhalten PROVIANTPRODUZIERENDEN ZAUBERMITTEL erhalten PROVOKATION Rache nehmen Rätsel lösen RÄTSELHAFTES Gespräch oder Handlung RÄTSELHAFTES Gespräch oder Handlung aufklären Rebellion des Zielobjektes VORTÄUSCHEN Reiche Erbschaft VORTÄUSCHEN Reichtum VORTÄUSCHEN Reitkünste zeigen REITTIER besorgen REITTIER erfassen REITTIER erhalten RENDEZVOUS RENDEZVOUS nachgehen RESPEKTLOSIGKEIT Reue VORTÄUSCHEN Richtiges auswählen Richtiges erlangen Sachschaden anrichten Sachverhalt aus der Vergangenheit aufklären Schatz entdecken Schicksal abwenden Schicksalsbestimmender Weisung folgen SCHLAF SCHLAF unterbrechen Schlußfolgerung als ABSURDITÄT Schrecken ausgelöst SCHWEIGEPFLICHT beheben SCHWEIGEPFLICHT verhängen SELTENEN Gegenstand besorgen SELTENEN Gegenstand erfassen Sich abseilen LASSEN Sich auf Eindringlinge einlassen Sich einem magischen EINGRIFF unterziehen Sich niederlassen Sich oder Andere hochziehen LASSEN Sich oder andere tragen LASSEN Sich sonstige Dinge ANEIGNEN Sich ZAUBERMITTEL ANEIGNEN Sich ZURÜCKZIEHEN SONSTIGE AUFGABE erledigen SONSTIGE AUFGABE in Angriff nehmen Sonstiges herstellen Sonstiges kaufen Sonstiges verkaufen Sonstiges VORTÄUSCHEN Sorge um Freikaufsbedürftige Sorge um Hilfsbedürftige Sorge um Hungernde oder Durstige Sorge um Mitleidsbedürftige Sorge um Pflegebedürftige Sorge um Schutzbedürftige Stifter oder Helfer gewinnen Strafvollstreckenden ZAUBERMITTEL erhalten Streit schlichten Suizidversuch VORTÄUSCHEN TABU Behälter nicht öffnen TABU Feuer darf nicht erlöschen TABU Frist nicht überziehen TABU Gegenstand nicht weitergeben TABU Haus nicht verlassen dürfen TABU keine Eindringlinge einlassen TABU keine Pause machen TABU Land nicht betreten TABU nachgehen TABU nicht einschlafen dürfen TABU nicht küssen TABU Schweigegebot TABU sich von Etwas fernhalten TABU sich von Jemandem fernhalten TABU Sonstiges TABU Tier nicht töten TABU Tierhaut nicht verbrennen TABU Türe nicht öffnen dürfen Tapferkeit beweisen Tarnung gewährendes ZAUBERMITTEL erhalten Tier als Schutzgeist erhalten Tiere gegeneinander kämpfen lassen Tiere HÜTEN Tiere HÜTEN in Angriff nehmen Tiere MÄSTEN Tiere MÄSTEN in Angriff nehmen Tierjunges erhalten TIERKÖRPERTEIL besorgen TIERKÖRPERTEIL erfassen Tod von Angehörigen Tod VORTÄUSCHEN TODESAUFTRAG Töte zu lebenden VORTÄUSCHEN TÖTUNG TRAUM vom sozialen Aufstieg TRAUM wird wahr Treue VORTÄUSCHEN TREUEN HELFER erhalten TREUEN HELFER finden Truhe herstellen Turm Zaun oder Berg hinaufsteigen Überraschende positive Wendung Überrascht werden oder geraten Um eine Frau werben Um einen Auftrag werben Um einen Mann werben Um GEGENLEISTUNG bitten Unerträgliche Faulheit Unerträgliche Schreckhaftigkeit Unerträgliche Unersättlichkeit Unerträgliche Ungeheuerlichkeit UNGEHEUER erfassen UNGEHEUER herbeiholen Ungewollte Verlobung akzeptieren Ungewünschtes Wesen beseitigen Universellen ZAUBERMITTEL erfassen Universellen ZAUBERMITTEL erhalten Unterkommen Unwissenheit oder Unsicherheit VORTÄUSCHEN Unzufriedenheit VORTÄUSCHEN Ursache des Altaussehens deuten Ursache des unstillbaren Dursts deuten Ursache des unstillbaren Hungers deuten VAMPIRISMUS VAMPIRISMUS beheben VERBANNUNG VERBANNUNG beheben Verbot als ABSURDITÄT Verdreht interpretieren VERFOLGUNG VERFOLGUNG entkommen VERFOLGUNG stören VERLEUMDUNG beheben VERLEUMDUNG bei Kindstötung VERLEUMDUNG bei Missgeburt VERLEUMDUNG bei Schädigung VERLEUMDUNG bei Untreue Verlobung wird arrangiert VERLOCKUNG VERLUST von Tieren Früchten Ernte usw Verlust VORTÄUSCHEN VERLUSTursache beheben VERMISSEN der Angehörige VERMISSEN der Angehörige beheben Vermittlung erhalten VERSCHLINGEN VERSCHLINGEN beheben VERSPRECHEN im Austausch für Hilfe Verstecken als Rettung VERSTECKWETTE VERSTECKWETTE in Angriff nehmen VERSTEINERUNG VERSTEINERUNG beheben VERTRAG abschließen oder verkünden VERTRAG auflösen VERTRAGsbedingungen erfüllen Vertrauen beweisen VERTREIBUNG VERTREIBUNG beheben VOGEL erfassen VOGEL herbeiholen Von einem AUSWEGLOSEN Ort entkommen VORHERSAGE VORHERSAGE erforderlich VORHERSAGE prüfen Wache halten Wachsamkeit oder Anwesenheit VORTÄUSCHEN Waffe Zaubermittel erfassen WAFFE ZAUBERMITTEL erhalten Wahl als ABSURDITÄT WASSERMANGEL WASSERMANGEL beheben WASSERQUELLE erfassen WASSERQUELLE herbeiholen Weg zum Paradies VORTÄUSCHEN Wegweisung erhalten Wegweisung erpressen Wegweisung mit ALTERNATIVEN WEIHE anordnen WEIHE vollziehen WEISHEITEN erhalten WEISHEITEN nachgehen Weiterverweisung an anderen Stifter oder Helfer Wette um die Verführung der Braut WETTKAMPF WETTKAMPF in Angriff nehmen WETTLAUF WETTLAUF in Angriff nehmen WIEDERBELEBEN WIEDERBELEBEN in Angriff nehmen Wiedergutmachung erpressen WILD besorgen WILD erfassen Wissen VORTÄUSCHEN Wunderbare EMPFÄNGNIS Wunderbare Transportmittel erhalten Wunsch erfüllendes ZAUBERMITTEL erhalten Zauberkunst erlernen Zauberkunstfertigkeit gewährleisten ZAUBERMITTEL AUSRÜSTUNG oder WAFFE ausprobieren ZAUBERMITTEL AUSRÜSTUNG oder WAFFE einsetzen ZOLLERHEBUNG in Angriff nehmen ZOLLERHEBUNG vollstrecken Zu Beauftragungen anregen Zug von Reisenden oder Karavane folgen Zum Schlafen bringen Zum STAUNEN bringen Zur Welt kommen ZURÜCKZIEHEN beheben Zusammenbringen aller Handlungsträger Zuschnappendes Hindernis überwinden ZWANGSEHE Drohung ZWANGSEHE Drohung beheben Zweikampf"


# Einzigartige Wörter extrahieren
result = extract_unique_words(text)

print("Einzigartige Wörter:", result)
```
